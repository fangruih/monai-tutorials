```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming the data is stored in a tab-separated text file
file_path = '/simurgh/u/fangruih/monai-tutorials/generative/3d_ldm/conditioning/fmriresults01.txt'  # Update this to the path of your text file

# Load the data
data = pd.read_csv(file_path, sep="\t")  # Using tab as a separator

# Extract the 'interview_age' column
ages = data['interview_age']
# Set global parameters for font sizes
plt.rcParams['font.size'] = 12  # Adjusts the base default size
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(ages, bins=20, color='blue', alpha=0.7, edgecolor='black')
plt.title('Histogram of Interview Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

```
